### Handel Missing Data in pandas

Topics Covered 
* fillna to fill missing  values using different ways
* interpolate to make a guess on missing values using interpolation
* dropna to drop rows with missing values

```python
import pandas as pd
url = "https://raw.githubusercontent.com/codebasics/py/master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv"
df = pd.read_csv(url)
df
```

for this we have to parse day column as date type

```python
df = pd.read_csv(url,parse_dates=['day'])
# set day as an index
df.set_index('day',inplace=True)
df
```

One way is to fill all the na value with some particular value using `fillna()` method

```python
new_df =df.fillna(0)
new_df
```

This above method is good for numerical data  but not for categorical data as in events column 0 doesn't make any sense.
For this we can use `modified fillna()`

```python
new_df =df.fillna({
    'temperature':0,
    'windspeed': 0,
    'event':"no event"
})
new_df
```

This looks good but this approach has also a problem
see the temperature column on day 1 the temp was 32 and on day 2 suddenly the temp drops to 0. this sounds a bit weird
also windspeed has alos a very little chance of being 0

Another approach is `forward fill`, means fill the missing value with previos day value

```python
new_df =df.fillna(method='ffill')
new_df
```

Same we can do backword fill

```python
new_df = df.fillna(method='bfill')
new_df
```

We can also do padding

```python
new_df = df.fillna(method='pad')
new_df
```

right now we are copying row wise i.e. we are pasting the above row value value for missing value but if our data is such that we need to fill the missing values side ways i.e. copying the column value we can use axis parameter

```python
new_df =df.fillna(method='bfill',axis='columns')
new_df
```

on 8th and 9th there are consecutive missing value ,on forward fill the value from 7th is copied twice, but lets say we for such cases we want to copy it only once for this we can use `limit` parameter

```python
new_df =df.fillna(method='ffill',limit=1)
new_df
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/codebasics/py/master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv"
df = pd.read_csv(url,parse_dates=['day'])
df.set_index("day",inplace=True)
df
```

The above methods of filling na element is good but not better as can be seen the temp on 1st is 32 and on 5th is 28 so temp on 4th must be in between 32 and 28 and not exactly the value of ffill or bfill.   fpr this pandas has `interpolate` method

```python
new_df = df.interpolate()
new_df
```

The above handeling of missing value makes much more sense but there is problem for 4th jan the value should be closer to 5th jan because the value for 2nd and 3rd jan is missing so we can interpolate using time

```python
new_df = df.interpolate(method='time')
new_df
```

Now the value for 4th is more closer to 5th which makes sense

Now what if I have to drop `na-values`

```python
new_df = df.dropna()
new_df
```

So it dropped all rows which had na values, but what if we have to drop only that row for which all columns have na value eg here for 9th jan there is  no value for any column

```python
new_df = df.dropna(how='all')
new_df
```

what if we have to drop all rows except that row which has two or in general n valid value for that we can use threshold value 

```python
new_df =df.dropna(thresh=2)
new_df
```

So this kept alll rows which has 2 valid values 

Now what if we have to fill for the missing days like 2nd and 3rd jan are missing, how to handel such cases

```python
df.reset_index(inplace=True)
dt = pd.date_range(df.day[0],df.day[len(df)-1])
df.set_index('day',inplace=True)
idx =pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```

### Merge DataFrame

```python
import pandas as pd
df1 = pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'temperature':[21,14,35]
})
df1
```

```python
df2 = pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'humidity':[65,68,75]
})
df2
```

```python
df3=pd.merge(df1,df2,on='city')
df3
```

Inner Join(Intersection) Outer join(Union)

```python
df1 = pd.DataFrame({
    'city':['new york','chicago','orlando','baltimore'],
    'temperature':[21,14,35,32]
})
df1
```

```python
df2 = pd.DataFrame({
    'city':['new york','chicago','san francisco'],
    'humidity':[71,74,75]
})
df2
```

if we do simple merge we get only common element i.e. inner join or intersection of the two df

```python
df3 = pd.merge(df1,df2,on='city')
df3
```

if we want to get all rows of both df we should use outer join

```python
df3 = pd.merge(df1,df2,on='city',how='outer')
df3
```

Some time we want to know from which df the data is coming for this use indicator


```python
df3 =pd.merge(df1,df2,on='city',how='outer',indicator=True)
df3
```

```python
df1 = pd.DataFrame({
    'city':['new york','chicago','orlando','baltimore'],
    'temperature':[21,14,35,32],
    'humidity':[71,74,75,78]
})
df1
```

right now it is adding suffix x and y we can cutomize suffix also
